```python
img = cv.imread("house.jpg", cv.IMREAD_COLOR)
cv2_imshow(img)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv2_imshow(img_rgb)
```

```python
img_gray = cv.cvtColor(img_rgb, cv.COLOR_RGB2GRAY)
cv2_imshow(img_gray)
```

```python
# PEAK BRIGHTNESS(ig)
contrast = 1
brightness = 100

adjusted_bright = cv.convertScaleAbs(img_gray, beta=brightness)
# adjusted_bright = cv.convertScaleAbs(img_gray, alpha=contrast, beta=brightness)
cv2_imshow(adjusted_bright)
```

```python
# PEAK DARKNESS(?)
contrast = 0.5
brightness = 0
# adjusted_dark = cv.convertScaleAbs(img_gray, beta=brightness)
adjusted_dark = cv.convertScaleAbs(img_gray, alpha=contrast, beta=brightness)
cv2_imshow(adjusted_dark)
```

```python
thresh = 127
img_gray_bw = cv.threshold(img_gray, thresh, 255, cv.THRESH_BINARY)[1]
cv2_imshow(img_gray_bw)
```

```python
number_of_white_pix = np.sum(img_gray_bw == 255)
number_of_black_pix = np.sum(img_gray_bw == 0)
  
print('Number of white pixels:', number_of_white_pix)
print('Number of black pixels:', number_of_black_pix)
# cv.countNonZero(img_gray_bw)

height = np.size(img_gray_bw, 0)
width = np.size(img_gray_bw, 1)
print("Width:", width)
print("Height:", height)
print("Total pixels:", width*height)
```
